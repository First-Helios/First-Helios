## Metadata scan — all files

```python
def scan_dta(filename):
    path = os.path.join(DATA_DIR, filename)
    df, meta = pyreadstat.read_dta(path)
    print(f"\n{'='*60}")
    print(f"FILE: {filename}")
    print(f"  Rows: {len(df):,}   Cols: {len(df.columns)}")
    print(f"  Dataset label: {meta.file_label or '(none)'}")
    if meta.notes:
        print(f"  Notes: {meta.notes}")
    print(f"\n  COLUMNS ({len(df.columns)}):")
    for col in df.columns:
        label = meta.column_labels[meta.column_names.index(col)] if col in meta.column_names else ""
        dtype = str(df[col].dtype)
        n_unique = df[col].nunique()
        vl = list(meta.variable_value_labels.get(col, {}).items())[:3]
        vl_str = f"  val_labels: {dict(vl)}{'...' if len(meta.variable_value_labels.get(col,{}))>3 else ''}" if vl else ""
        print(f"    {col:<30} {dtype:<10} {n_unique:>6} unique  {label}{vl_str}")
    return df, meta

# cp-meta-publicdataset.dta and Trajectories-10-years-dataset.dta are unreadable (corrupt/format issue)

all_data = {}
for fname in sorted(dta_files):
    df, meta = scan_dta(fname)
    all_data[fname] = (df, meta)
```

## "Next Industry" Recommendation Engine
#
# Data sources used:
#   Dashboard-transitions-dataset.dta  — actual occupation-to-occupation transition flows
#                                         with wage change direction (up/lateral/down)
#   Dashboard-trajectories-dataset.dta — 3/5yr wage growth trajectories per occupation
#   Emsi-dataset.dta                   — job zone (training required), starting wage, regional demand
#
# Logic:
#   Given a starting occupation (SOC code or title), find the top transitions where:
#     1. TransitionOrder is low (most common moves)
#     2. TransitionWageDirection >= 0 (lateral or upward wage move)
#     3. Enriched with trajectory wage growth at 3yr and 5yr
#     4. Enriched with Emsi job zone and starting wage

```python
import pyreadstat
import pandas as pd

DATA_DIR = "/home/fortune/Downloads/PublicPoolData/Employment/"

# Load the three key datasets
df_trans, _ = pyreadstat.read_dta(DATA_DIR + "Dashboard-transitions-dataset.dta")
df_traj,  _ = pyreadstat.read_dta(DATA_DIR + "Dashboard-trajectories-dataset.dta")
df_emsi,  _ = pyreadstat.read_dta(DATA_DIR + "Emsi-dataset.dta")

print(f"Transitions:  {df_trans.shape}  cols: {list(df_trans.columns)}")
print(f"Trajectories: {df_traj.shape}  cols: {list(df_traj.columns[:10])}...")
print(f"Emsi:         {df_emsi.shape}  cols: {list(df_emsi.columns)}")
```

```python
# Build enriched transitions reference table
# Emsi has: origin SOC, dest SOC, median wages, skill deltas, transition order, wage direction
# Trajectories has: CensusCode → starting wage percentile, 3/5yr wage growth

# Trajectory summary: one row per (CensusCode, YearAfterStart) → collapse to per-occupation
# Use 3yr and 5yr rows from Dashboard-trajectories
df_traj3, _ = pyreadstat.read_dta(DATA_DIR + "Trajectories-3-years-dataset.dta")
df_traj5, _ = pyreadstat.read_dta(DATA_DIR + "Trajectories-5-years-dataset.dta")

# Get median wage growth and pct earning >$25/hr per CensusCode
# These are individual-level records — aggregate to occupation level
def agg_trajectories(df):
    return (
        df.groupby("CensusTitle")
        .agg(
            med_wage_growth=("WtMedWageGrowth", "median"),
            pct_earn_25plus=("PctWage25ormore", "median"),
            start_wage_pctile=("StartWagePctle", "median"),
            med_start_wage=("WtMedStartWage", "median"),
        )
        .reset_index()
    )

traj3_agg = agg_trajectories(df_traj3)
traj5_agg = agg_trajectories(df_traj5)

print(f"Traj 3yr occupations: {len(traj3_agg)}")
print(f"Traj 5yr occupations: {len(traj5_agg)}")
display(traj3_agg.head())
```

```python
def recommend_next_occupation(
    occupation: str,
    top_n: int = 10,
    wage_filter: str = "any",   # "up", "lateral_or_up", "any"
    max_skill_gap: float = None, # optional: max absolute avg skill gap (lower = easier transition)
):
    """
    Given a starting occupation title or SOC code, return the top recommended
    next occupations based on:
      1. Actual transition frequency (order)
      2. Wage direction (up / lateral / down)
      3. 3yr and 5yr wage growth trajectory at the destination
      4. Skill transferability (avg ISA skill gap from Emsi)

    Parameters
    ----------
    occupation   : str  — partial match on occ_title_origin or oesCode_origin
    top_n        : int  — max recommendations to return
    wage_filter  : "up" | "lateral_or_up" | "any"
    max_skill_gap: float — optional ceiling on avg absolute skill difference
    """
    isa_cols = [c for c in df_emsi.columns if c.startswith("isa")]

    # --- Match source occupation ---
    mask = df_emsi["occ_title_origin"].str.contains(occupation, case=False, na=False)
    if not mask.any():
        mask = df_emsi["oesCode_origin"].astype(str).str.startswith(occupation)
    if not mask.any():
        print(f"No match found for '{occupation}'")
        print("Available origins (sample):", df_emsi["occ_title_origin"].unique()[:10].tolist())
        return None

    subset = df_emsi[mask].copy()
    origin_title = subset["occ_title_origin"].iloc[0]
    origin_soc   = subset["oesCode_origin"].iloc[0]
    origin_wage  = subset["h_median_origin"].iloc[0]
    print(f"Source: {origin_title} (SOC {origin_soc})  |  Median hourly: ${origin_wage:.2f}")

    # --- Wage direction filter ---
    if wage_filter == "up":
        subset = subset[subset["category"] == 1.0]
    elif wage_filter == "lateral_or_up":
        subset = subset[subset["category"] >= 0.0]

    # --- Skill gap: avg absolute ISA delta ---
    subset = subset.copy()
    subset["avg_skill_gap"] = subset[isa_cols].abs().mean(axis=1)
    subset["skill_gap_direction"] = subset[isa_cols].mean(axis=1)  # positive = destination needs more skill

    if max_skill_gap is not None:
        subset = subset[subset["avg_skill_gap"] <= max_skill_gap]

    # --- Sort: frequency first, then wage direction, then skill gap ---
    direction_label = {1.0: "↑ Up", 0.0: "→ Lateral", -1.0: "↓ Down"}
    subset = subset.sort_values(["order", "avg_skill_gap"]).head(top_n * 3)

    # --- Enrich with 3yr trajectory at destination ---
    subset = subset.merge(
        traj3_agg.rename(columns={
            "CensusTitle": "occ_title_dest",
            "med_wage_growth": "dest_wage_growth_3yr",
            "pct_earn_25plus": "dest_pct_25plus_3yr",
        })[["occ_title_dest", "dest_wage_growth_3yr", "dest_pct_25plus_3yr"]],
        on="occ_title_dest",
        how="left",
    )

    # --- Build output table ---
    out_cols = [
        "occ_title_dest", "occ_family_dest",
        "h_median_dest", "med_wage_diff",
        "category", "PctUp", "PctLateral", "PctDown",
        "avg_skill_gap", "skill_gap_direction",
        "dest_wage_growth_3yr", "dest_pct_25plus_3yr",
        "lic_new",
    ]
    result = subset[out_cols].head(top_n).copy()
    result["wage_move"] = result["category"].map(direction_label)
    result["dest_pct_25plus_3yr"] = (result["dest_pct_25plus_3yr"] * 100).round(1).astype(str) + "%"
    result["dest_wage_growth_3yr"] = result["dest_wage_growth_3yr"].map(lambda x: f"+${x:.2f}/hr" if pd.notna(x) else "n/a")
    result["med_wage_diff"] = result["med_wage_diff"].map(lambda x: f"+${x:.2f}" if x >= 0 else f"-${abs(x):.2f}")
    result["h_median_dest"] = result["h_median_dest"].map(lambda x: f"${x:.2f}")
    result["avg_skill_gap"] = result["avg_skill_gap"].round(3)
    result["needs_license"] = result["lic_new"].map({1: "Yes", 0: "No"})

    display(result[[
        "occ_title_dest", "wage_move", "h_median_dest", "med_wage_diff",
        "dest_wage_growth_3yr", "dest_pct_25plus_3yr",
        "avg_skill_gap", "needs_license",
    ]].rename(columns={
        "occ_title_dest":       "Destination Occupation",
        "wage_move":            "Wage Move",
        "h_median_dest":        "Median Hourly",
        "med_wage_diff":        "Wage Change",
        "dest_wage_growth_3yr": "3yr Wage Growth",
        "dest_pct_25plus_3yr":  "% Earn $25+/hr (3yr)",
        "avg_skill_gap":        "Avg Skill Gap",
        "needs_license":        "New License?",
    }))
    return result
```

```python
## Example: Food prep worker → lateral or upward moves
print("=== Top transitions from food prep/service roles ===\n")
recommend_next_occupation("Food Preparation", wage_filter="lateral_or_up", top_n=10)
```

```python
## Example: Cashier / retail → upward only, low skill gap
print("=== Top upward transitions from cashiers (low skill gap) ===\n")
recommend_next_occupation("Cashier", wage_filter="up", max_skill_gap=0.3, top_n=10)
```
